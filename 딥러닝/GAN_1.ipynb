```python
# 필요한 라이브러리를 로드합니다
from tensorflow.keras.datasets import mnist
from tensorflow.keras.layers import Input, Dense, Dropout
from tensorflow.keras.models import Sequential, Model
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt

# 출력과 노이즈를 지정합니다
n_output = 784
n_noise = 128

generator = Sequential() # 생성자 모델을 선형으로 적재하기 위해 Sequential()로 선언합니다
generator.add(Dense(units=256, input_dim=n_noise, activation='relu')) # 유닛 수가 256, 입력 차원은 입력 노이즈 차원으로, 활성화 함수를 relu로 하는 완전연결층을 추가합니다
generator.add(Dense(units=512, activation='relu')) # 유닛 수가 512, 활성화 함수로 relu를 사용하는 완전연결층을 추가합니다
generator.add(Dense(units=n_output, activation='tanh')) # 유닛 수는 출력될 이미지 차원으로, 활성화 함수로 tanh를 사용하는 완전연결층을 추가합니다

discriminator = Sequential() # 판별자 모델을 선형으로 적재하기 위해 Sequential()로 선언합니다
discriminator.add(Dense(256, input_dim=n_output, activation='relu')) # 출력 차원이 256, 입력 차원은 출력될 이미지 차원으로, 활성화 함수로 tanh를 사용하는 완전연결층을 추가합니다
discriminator.add(Dropout(0.3)) # 0.3의 비율로 드롭아웃을 수행합니다
discriminator.add(Dense(1, activation='sigmoid')) # 이진 분류를 수행하기 위해 출력이 1이고 활성화 함수로 sigmoid를 사용하는 완전연결층을 추가합니다

g_input = Input(shape=(n_noise,)) # Input 함수를 사용하여 n_noise 차원의 입력 노이즈를 g_input으로 생성합니다
g_output = discriminator(generator(g_input)) # 생성자 모델에 g_input을 넣어 가짜 이미지를 생성 후 판별자 모델에 입력하여 예측값을 g_output으로 얻습니다
gan = Model(g_input, g_output) # Model 함수를 사용하여 g_input이 g_output으로 되는 과정을 연결해 GAN모델을 생성합니다

discriminator.trainable = True # 판별자를 학습 가능하게 설정합니다
adam = tf.optimizers.Adam(lr=0.0002, beta_1=0.5) # learning rate가 0.0002인 Adam을 선언합니다
discriminator.compile(loss='binary_crossentropy', optimizer='adam') # 손실함수로 이진 crossentropy를, 옵티마이저로 위에서 선언한 Adam을 사용하는 판별자를 컴파일합니다

discriminator.trainable = False # 이후 판별자를 학습 불가능하게 설정하여 GAN 모델 컴파일시 생성자만 학습합니다
gan.compile(loss='binary_crossentropy', optimizer='adam') # 손실함수로 이진 crossentropy를, 옵티마이저로 위에서 선언한 Adam을 사용하는 GAN모델을 컴파일합니다

gan.summary() # 모델의 정보를 요약합니다

# 학습을 위한 요소를 설정합니다
n_epoch = 50
batch_size = 128
saving_interval = 10


(train_images, _), (_, _) = mnist.load_data() # MNIST 데이터셋에서 훈련용 이미지만 불러옵니다  

buffer_size = len(train_images) # 버퍼 크기를 이미지의 수로 설정합니다
train_images = train_images.reshape((buffer_size, n_output)) # 훈련용 이미지를 (buffer_size, n_output) 차원으로 reshape합니다

train_images = (train_images - 127.5) / 127.5 # 이미지 픽셀 범위가 0~255이기 때문에 127.5를 빼고 나누어 -1 ~ 1로 정규화를 수행합니다

train_dataset = tf.data.Dataset.from_tensor_slices(train_images).shuffle(buffer_size).batch(batch_size) # 데이터를 랜덤하게 섞고 슬라이스합니다

# 학습을 진행합니다
for i in range(n_epoch):
    print('Epoch Num : {}/{}'.format((i + 1), n_epoch)) # 반복 횟수를 출력합니다

    for image_batch in train_dataset: # 이미지 배치 별로 반복합니다
        n_bt_imgs = len(image_batch) # 이미지 수는 이미지 배치의 길이입니다
        true_labels = np.ones((n_bt_imgs, 1)) # 이미지 배치 길이를 가지며 1로 초기화된 진짜 레이블을 생성합니다
        fake_labels = np.zeros((n_bt_imgs, 1)) # 이미지 배치 길이를 가지며 0으로 초기화된 가짜 레이블을 생성합니다

        d_loss_real = discriminator.train_on_batch(image_batch, true_labels) # 이미지 배치와 정답 레이블을 이용해 판별자로 훈련하고 손실을 획득합니다

        noise = np.random.normal(0, 1, (n_bt_imgs, n_noise)) # 정규분포에서 (n_bt_imgs, n_noise)크기의 노이즈를 생성합니다
        gen_imgs = generator.predict(noise) # 노이즈를 입력으로 생성자 모델을 사용해 가짜 이미지를 생성합니다
        d_loss_fake = discriminator.train_on_batch(gen_imgs, fake_labels) # 가짜 이미지와 가짜 레이블을 이용해 판별자로 훈련하고 손실을 획득합니다
        d_loss = 0.5 * np.add(d_loss_real, d_loss_fake) # 진짜 이미지에 대한 손실과 가짜 이미지에 대한 손실에 대한 평균을 구합니다
      
        g_loss = gan.train_on_batch(noise, true_labels) # 노이즈와 정답 레이블을 사용해 생성자로 훈련하고 손실을 획득합니다

    # 조건에 맞는 경우 수치들을 출력해줍니다 
    if (i +1) % saving_interval == 0 or i == 1:
        print('Epoch:%d' % (i + 1), 
             'd_loss:%.4f' % d_loss, ' g_loss:%.4f' % g_loss)
        
        noise = np.random.normal(0, 1, (25, n_noise)) # 정규분포에서 (25, n_noise)크기의 노이즈를 생성합니다
        gen_imgs = generator.predict(noise) # 노이즈를 사용해 생성자로 예측합니다

        #gen_imgs = 0.5 * gen_imgs + 0.5  
        
        # 예측한 이미지를 플로팅하고 별도로 저장하는 작업입니다
        fig, axs = plt.subplots(5, 5)  
        count = 0
        for j in range(5):
            for k in range(5):
                axs[j, k].imshow(np.reshape(gen_imgs[count], (28, 28)))
                axs[j, k].axis('off')
                count += 1
        plt.show()
        fig.savefig("gan_results/g_mnist_%d.png" % (i + 1))
```
